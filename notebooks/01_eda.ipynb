```python
import pandas as pd

train = pd.read_csv('../data/raw/train.csv')
test = pd.read_csv('../data/raw/test.csv')

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("\nColumns:", train.columns.tolist())
```

```python
print("Train columns:", train.columns.tolist())
print("\nTest columns:", test.columns.tolist())

print("\nMissing in test:", set(train.columns) - set(test.columns))

```

```python
print("Emotional State Classes:")
print(train['emotional_state'].value_counts())

print("\nIntensity Values:")
print(train['intensity'].value_counts().sort_index())

print("\nTotal unique emotional states:", train['emotional_state'].nunique())
print("Total unique intensity values:", train['intensity'].nunique())
```

```python
print("Missing values in Train:")
print(train.isnull().sum())

print("\nMissing values in Test:")
print(test.isnull().sum())
```

```python
import pandas as pd

train = pd.read_csv('../data/raw/train.csv')
test = pd.read_csv('../data/raw/test.csv')

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("\nColumns:", train.columns.tolist())
```

```python
print("Train columns:", train.columns.tolist())
print("\nTest columns:", test.columns.tolist())

print("\nMissing in test:", set(train.columns) - set(test.columns))

```

```python
print("Emotional State Classes:")
print(train['emotional_state'].value_counts())

print("\nIntensity Values:")
print(train['intensity'].value_counts().sort_index())

print("\nTotal unique emotional states:", train['emotional_state'].nunique())
print("Total unique intensity values:", train['intensity'].nunique())
```

```python
print("Missing values in Train:")
print(train.isnull().sum())

print("\nMissing values in Test:")
print(test.isnull().sum())
```

```python
train['text_length'] = train['journal_text'].str.len()
train['word_count'] = train['journal_text'].str.split().str.len()

print("Text Length Stats:")
print(train['text_length'].describe())

print("\nWord Count Stats:")
print(train['word_count'].describe())

print("\nShortest 5 journal entries:")
print(train.nsmallest(5, 'word_count')[['journal_text', 'word_count', 'emotional_state']])
```

```python
print("Shortest 5 journal entries:")
print(train.nsmallest(5, 'word_count')[['journal_text', 'word_count', 'emotional_state']])
```

```python
print("Ambience Types:")
print(train['ambience_type'].value_counts())

print("\nTime of Day:")
print(train['time_of_day'].value_counts())

print("\nPrevious Day Mood:")
print(train['previous_day_mood'].value_counts())

print("\nFace Emotion Hint:")
print(train['face_emotion_hint'].value_counts())

print("\nReflection Quality:")
print(train['reflection_quality'].value_counts())
```

```python
print("Numerical Features Summary:")
print(train[['sleep_hours', 'energy_level', 'stress_level', 'duration_min']].describe())

print("\nEnergy Level values:")
print(train['energy_level'].value_counts().sort_index())

print("\nStress Level values:")
print(train['stress_level'].value_counts().sort_index())
```
